```python
# start by importing our libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm
from matplotlib import animation
from IPython.display import HTML
import matplotlib.cm as cm

# now import functions we are going to use
from test_functions import *
from ga import *
from lm import *
from leapfrog import *
from isd import *
from hooke_jeeves import *
from hga import *

# add our functions and optimizers
functions = [ackley, beale, booth, bukin,easom, goldstein, holder, matyas, rosenbrock, sphere]
fun_names = ['Ackley', 'Beale', 'Booth', 'Bukin', 'Easom',
             'Goldstein', 'Holder', 'Matyas', 'rosenbrock',  'Sphere']

opt_names = ['HJ', 'ISD', 'LF', 'LM', 'GA', 'HGA LF', 'HGA HJ', 'HGA LM']
optimizers = [hooke_jeeves, isd, leapfrog, lm, ga, hga_lf, hga_hj, hga_lm]
```

```python
# Plotting the each function 

for idx, f in enumerate(functions):
    print(idx+1, " of ", len(functions)) 
    xmin, xmax, ymin, ymax = get_range(f)
    real_min = get_final_min(f)
    real_coords = get_final_coords(f)
    xstep = .2
    ystep = .2
    x, y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep), np.arange(ymin, ymax + ystep, ystep))
    z = f(x,y)
    
    # get points for each optimizer and store in dictionary
    points = all_mins ={}
    for opt, name in zip(optimizers, opt_names):
        
        tmin, xy = opt(f,verbose=True)
        
        points[name] = (xy)
        #all_mins[name] = tmin
    
        
        
    # plot 3d 
    fig = plt.figure(figsize=(8,8))
    ax = plt.axes(projection='3d')
    ax.set_xlim([xmin,xmax])
    ax.set_ylim([ymin,ymax])
    ax.plot_surface(x, y, z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
    ax.set_title(fun_names[idx], weight='heavy', fontsize=20)
    
    # plot contour
    fig, cs = plt.subplots(figsize=(8, 8))
    cs.set_xlim([xmin,xmax])
    cs.set_ylim([ymin,ymax])
    cs.contour(x, y, z)
    cs.plot(real_coords[0], real_coords[1], 'k*', markersize=10)
    
    # plot the best points of each function
    colors = cm.rainbow(np.linspace(0, 1, len(optimizers))).tolist()
    for keys, pts in points.items():
        
        px, py = zip(pts) # unpack
        cs.plot(px, py, label=keys,marker='+', color=colors.pop(),markersize=15, mew=1.75)
        idx += 1
        
    box = cs.get_position()
    cs.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    cs.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
print('Done.')
```
